Lista a todos

```python
lista = [1,2,3,3]
print(lista)
print("1. Tupla:", tuple(lista), type(tuple(lista)))
print("2. Conjunto", set(lista), type(set(lista)))
print("3. Diccionario", dict(enumerate(lista)), type(dict(enumerate(lista))))
```
