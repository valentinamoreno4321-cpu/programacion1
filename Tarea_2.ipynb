```python
# Conjunto

print(set())

print(set([5, 2, 5, 1, 1.5]))
print(set((5, 2, 5, 1, 1.5)))
print(set("52511.5"))

conjunto = set([2, 3, 3, 4])
conjunto_2 = set([5, 3, 5, 6])
conjunto_3 = set([4, 2])

print(conjunto)
print(conjunto_2)
print(conjunto_3)

conjunto.add(1)
print(conjunto)

conjunto.remove(1)
print(conjunto)

print(conjunto.intersection(conjunto_2))
print(conjunto_2.issubset(conjunto))
print(conjunto_3.issubset(conjunto))
```

```python
# Diccionario

diccionario = {1: "Uno", 2: "Dos"}
diccionario[3] = "Tres"
print(diccionario)

dict_lista_tuplas = dict([(1, "Uno"), (2, "Dos"), (3, "Tres")])
print(dict_lista_tuplas)

dict_lista_string = dict(Uno = 1, Dos = 2, Tres = 3)
print(dict_lista_string)

dict_tipos = {1: "integer",
              2.2: "float",
              "texto": "string",
              (1, 2): "tupla"}

print(dict_tipos)

dict_repeticion = {1: "Primero", 1: "Último"}
print(dict_repeticion)

print(diccionario,
      diccionario.keys(),
      diccionario.values(),
      diccionario.items())

claves = diccionario.values()
print(claves)
diccionario[1] = "One"
print(claves)
diccionario.pop(2)
print(diccionario)
```

```python
# Ejemplo de conversión de distintos tipos a False

print(0, "->", bool(0))
print(0.0, "->", bool(0.0))
print("", "->", bool(""))
print(list(), "->", bool(list()))
print(tuple(), "->", bool(tuple()))
print(set(), "->", bool(set()))
print(dict(), "->", bool(dict()))
print(None, "->", bool(None))
```

```python
# Ejemplo de conversión de distintos tipos a True

print(9000, "->", bool(9000))
print(3.1415, "->", bool(3.1415))
print("Cadena", "->", bool("Cadena"))
print(list((1, 2)), "->", bool(list((1, 2))))
print(tuple((1, 2)), "->", bool(tuple((1, 2))))
print(set((1, 2)), "->", bool(set((1, 2))))
print(dict([(1,2)]), "->", bool(dict([(1,1)])))
```

```python
# Ejemplo if-else

estrato = float(input("Introduzca su estrato: "))
if estrato < 3:
  print("Puedes obtener el subsidio, Felicidades")
else:
  print("No puedes obtener el subsidio")
```

```python
#Ejemplo if-elif-else

estrato = float(input("Introduzca su estrato: "))
if 1 <= estrato and estrato < 2:
  print("Se encuentran en una situación de extrema pobreza.")
elif 2 <= estrato and estrato < 3:
  print("Se encuentran en una situación de pobreza.")
elif 3 <= estrato and estrato < 4:
  print("Se encuentran en una situación de vulnerabilidad.")
elif 4 <= estrato and estrato < 5:
  print("Se encuentran en una situación de no pobreza.")
elif 5 <= estrato and estrato < 6:
  print("Se encuentran en una situación de no pobreza y no vulnerabilidad.")
else:
  print("Estrato fuera de rango")
```

```python
# Ejemplo while

frutas = 5
while frutas != 0:
  frutas = frutas - 1
  print("La cantidad de frutas es", frutas)
```

```python
# Ejemplo básico del for

for colores in ["rojo", "amarillo", "verde"]:
  print(colores)
```

```python
# for sobre estructuras de datos

lista = list(("rojo", "amarillo", "verde"))
tupla = tuple(("rojo", "amarillo", "verde"))
conjunto = set(("rojo", "amarillo", "verde"))
diccionario = dict(((2,"rojo"), (1, "amarillo"), (0, "verde")))

for colores in lista:
  print(colores)
for colores in diccionario:
  print(colores)
```

```python
# Acceso por índices a una lista con un for

lista = ["rojo", "amarillo", "verde"]

for indice in range(len(lista)):
  print("Indice:", indice,"Elemento:", lista[indice])
```

```python
# for para visualizar el efecto de las sentencias break y continue

for numero in range(15):
  print(numero)
```

```python
# Visualización del efecto de la sentencia break

for numero in range(15):
  if numero == 3:
    break
  print(numero)
```

```python
# Visualización del efecto de la sentencia continue

for numero in range(15):
  if numero == 3:
    continue
  print(numero)
```

```python
# Uso del else en un while

intentos = 4
while intentos > 0:
  if input("Escriba la contraseña:") == "Vale123":
    print("¡Correcta!")
    break
  intentos = intentos - 1
  print("Contraseña incorrecta")
else:
  print("Se te acabaron los intentos :(")
```
