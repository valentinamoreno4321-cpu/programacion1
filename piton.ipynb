```python
# Series de Fibonacci:
# la suma de dos elementos define el siguiente
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b # El valor de b no está actualizado, por tanto toma el valor anterior
```

```python
a, b = 0, 1
while b < 100:
  print(b)
  a = b
  b = a+b # Le da tiempo a que el valor de b se actualice
```

```python
a, b = 0, 1
for b in range(20):   # 6 repeticiones
    print(b)
    a, b = b, a + b
```

```python
x = int(input("Ingresa un entero, por favor: "))
if x < 0:
  x = 0
  print('Negativo cambiado a cero')
elif x == 0:
  print('Cero')
elif x == 1:
    print('Uno')
else:
  print('Mayor que uno')
```
